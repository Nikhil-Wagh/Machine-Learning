# Linear Regression Gradient Descent
### 1. Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 2. Sampling Data
###### a. Generating Data
Noise is not generated in the training data set so as to learn quickly. With noise the GRADIENT DESCENT *overshoots* the minimum.
Althought it was introduced in the test data set.

```python
no_of_samples = 20
no_of_test_samples = 10
train_x = np.linspace(0, 20, no_of_samples)
train_y = -2*train_x + 5 
test_x = np.linspace(30, 40, no_of_test_samples)
test_y = -2*test_x + 5 + (np.random.random(test_x.shape)*10-5)
```

###### b. Plotting data to get a visual

```python
plt.scatter(train_x, train_y, color = "red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Fig 1. Data for Linear Regression Gradient Descent")
plt.show()
```

### 3. Training the machine
To train the machine using Gradient descent *alpha* is taken as 0.01 and 100 iterations over the training data set are done.
Any line in the form of **Y = Ax + B** can be drawn with the help of the variables **A** and **B** .
In Gradient Descent we start by assuming any random value of features and apply gradient descent to reach the local minima.
To find A and B we need to find the error on each iteration.
value of B at any iteration t could be found as follows:
`B(t) = B(t-1) - alpha*error`
similarly the value of A at any iteration could be found as follows:
`A(t) = A(t-1) - alpha*error*x_value_of_that_iteration`
These equations are found by partial derivation of the function *Y = Ax + B* with respect to the features A and B, respectively.
###### a. Training and plotting the predicted data of each iteration

```python
A = 0
B = 0
save_B = [0]
save_A = [0]
alpha = 0.01
count=0
colors = ['red', 'green', 'orange', 'yellow']   
for i in range (0, 100):
    for j in range(0, len(train_x)):
        predicted_y = A*train_x[j] + B
        error = predicted_y - train_y[j]  
        B = B - alpha*error
        A = A - alpha*error*train_x[j]
        save_B.append(B)
        save_A.append(A)
        plt.scatter(train_x[j], predicted_y,color=colors[i%4])
plt.plot(train_x, train_y, c="blue")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Fig 2. Predicted and actual data comparison")
plt.show()
```

###### b. Change in value of B with each iteration

```python
plt.figure()
plt.plot(save_B,color="purple")
plt.xlabel("Number of Iterations")
plt.ylabel("Value of B")
plt.title("Fig 3. Change in value of B")
plt.show()
```

###### c. Change in value of A with each iteration

```python
plt.figure()
plt.plot(save_A,color="purple")
plt.xlabel("Number of Iterations")
plt.ylabel("Value of A")
plt.title("Fig 3. Change in value of A")
plt.show()
```

### 4. Predicting data for Test Dataset

```python
test_predicted_y = A*test_x + B
plt.scatter(test_x, test_predicted_y, color="red")
plt.scatter(test_x, test_y, color="blue")
plt.show()
```

### 5. Mean squared Error on the test Dataset

```python
print("MSE: ", np.mean((test_predicted_y - test_y)**2))
```
