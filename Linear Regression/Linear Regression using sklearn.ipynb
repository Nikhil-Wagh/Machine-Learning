```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
no_of_samples = 100
x = np.linspace(-np.pi, np.pi, no_of_samples)
y =  0.5*x + np.sin(x) + np.random.random(x.shape)
plt.scatter(x, y, color="red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Fig 1. Sampling data for Linear Regression")
plt.show()
```

```python
random_indices = np.random.permutation(no_of_samples)
train_x = x[random_indices[:70]]
train_y = y[random_indices[:70]]
validation_x = x[random_indices[70:85]]
validation_y = y[random_indices[70:85]]
test_x = x[random_indices[85:]]
test_y = y[random_indices[85:]]
```

```python
model = linear_model.LinearRegression()
train_x_for_fitting = np.matrix(train_x.reshape(len(train_x),1))
train_y_for_fitting = np.matrix(train_y.reshape(len(train_y),1))

model.fit(train_x_for_fitting, train_y_for_fitting)

# predicted_train_y = model.predict(train_x_for_fitting)
# predicted_validation_y = model.predict(validation_x.reshape(len(validation_x),1))
# predicted_test_y = model.predict(test_x.reshape(len(test_x),1))

# plt.plot(train_x_for_fitting, predicted_y, color="blue")
# plt.plot(validation_x, predicted_validation_y, color = "orange")
# plt.plot(test_x, predicted_test_y, color="yellow")

plt.scatter(x, y, color="red")

plt.plot(x, model.predict(x.reshape(len(x),1)), color="blue")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Fig 2. Predicted Data")
plt.show()
```

```python
print("MSE Complete Set: ", np.mean((y - model.predict(x.reshape(len(x),1)))**2))
print("MSE Validation Set: ", np.mean((validation_y - model.predict(validation_x.reshape(len(validation_x),1)))**2))
print("MSE Test Set: ", np.mean((test_y - model.predict(test_y.reshape(len(test_y),1)))**2))
```
