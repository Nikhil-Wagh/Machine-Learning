# Linear Regression
### 1. Imports

```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np
```

### 2. Sampling of data
Here the function used to sample the data is ` y = -2x + 5 ` is in the form ` Y = Ax + B` where A and B are constants.
Sampling is done by using *Gaussian Noise* which belongs to (0,1).
` (np.random.random(x.shape)*10-5) ` is done to introduce noise to the data.

```python
no_of_samples = 100
x = np.linspace(-10, 10, no_of_samples)
y = -2*x+5+(np.random.random(x.shape)*10-5)
plt.scatter(x,y,color="red")    
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Fig 1: Data for linear regression')
plt.show()
```

### 3. Predict data using Differentiation
To minimize the **Cost Function** *J=sum((y-H(x))^2)* we use differentiation and differentiate the cost function and equate it with 0, to get the value of constants (in our case A and B)
For the line of the form Y = Ax +B
The values of A and B after differentiation is found to be:
`A = (n*sum(x*y) - sum(x)*sum(y))/(nsum(x^2)-(sum(x))^2)
B = (sum(y) - A(sum(x)))/n `

```python
A = (no_of_samples*sum(x*y) - sum(x)*sum(y))/(no_of_samples*sum(np.square(x))-(sum(x))**2)
B = (sum(y) - A*(sum(x)))/no_of_samples
print(A, B)
```

```python
y_predict = A*x + B
```

### 4. Plot the Predicted data and find the Mean Squared Error on the training set

```python
plt.scatter(x, y, color="red")
plt.plot(x, y_predict, color="blue")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Fig2. Predicted data")
plt.show()
```

```python
print("MSE: ", sum(np.sqrt((y - y_predict)**2))/no_of_samples)
```

```python
def predict_y(x):
    predicted_y = A*x+B
    return predicted_y
```

### 5. Predict the output for Unknown Data
Sampling of test data is done in the similar manner like above.
And again the predicted data is plotted on the graph.
The Mean Squared error is also found in the similar manner.

```python
no_of_test_samples = 10
test_x = np.linspace(20, 30, no_of_test_samples)
test_y = -2*test_x+5+(np.random.random(test_x.shape)*10-5)
predicted_y = predict_y(test_x)
plt.scatter(test_x,test_y,color="red")
plt.plot(test_x, predicted_y, color="blue")
plt.show()
print("MSE: ", sum(np.sqrt((test_y-predicted_y)**2))/no_of_test_samples)
```
